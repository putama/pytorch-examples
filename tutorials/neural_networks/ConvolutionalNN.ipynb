## Convolutional NeuralNet implementation using Pytorch and Torchvision library
This example is using the MNIST database of handwritten digits
(http://yann.lecun.com/exdb/mnist/)

One important point that I want to show here is that how Pytorch nn.Module allows you to define your model on levels of hirarchy. You can wrap convolution and drop out layers into a sequential module, and wrap fully connected layers into another sequential module. You can later assemble them together into a single module on your forward() implementation. This is particularly useful to maintain the modularity of your model.

```python
import numpy as np

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.autograd import Variable

import torchvision as tv

import PIL
from PIL import Image

import matplotlib.pyplot as plt

%pylab inline
pylab.rcParams['figure.figsize'] = (10, 10)

import time
```

```python
# load the train images
# images are still in PIL image format
imgloader = tv.transforms.Compose([tv.transforms.ToTensor()])
imgunloader = tv.transforms.ToPILImage()
trainmnist = tv.datasets.MNIST('mnist/', train=True, download=True, transform=imgloader)
testmnist = tv.datasets.MNIST('mnist/', train=False, download=True, transform=imgloader)

# torch convenient tools to load manage dataset
batch_size = 100
trainloader = torch.utils.data.DataLoader(trainmnist, batch_size=batch_size, num_workers=4)
testloader = torch.utils.data.DataLoader(testmnist, batch_size=batch_size, num_workers=4)
```

```python
# visualize some of the first digit images
plt.subplots_adjust(hspace=.8)
for i in range(64):
    sbplt = plt.subplot(8,8,1+i)
    sbplt.set_title(trainmnist[i][1])
    sbplt.get_xaxis().set_visible(False)
    sbplt.get_yaxis().set_visible(False)
    plt.imshow(imgunloader(trainmnist[i][0]))
```

```python
# Training Parameters
learning_rate = 0.001
momentum = 0.9
training_epochs = 15
display_step = 1

# utilize cuda for computation
is_cuda = True
```

```python
# define the architecture of the convolutional network
class ConvNet(nn.Module):
    def __init__(self, num_classes):
        super(ConvNet, self).__init__()
        
        # filter and layer size is arbitrarily defined
        # you are free to experiment with this parameters
        self.num_classes = num_classes
        
        # keep in mind that mnist consist of single channel images
        # the convolutional layer would map the input from 
        # a single channel -> 32 channels -> 64 channels
        self.features_detector = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size=5, stride=1, padding=2),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(32, 64, kernel_size=5, stride=1, padding=2),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )
        
        # unfortunately, here there is no easy way to determine the size of
        # convolutional layer output to be the input of fully connected layer
        # we need to calculate the manually. in this case, we can be sure that
        # there are 64 output channel from the previous layer. therefore, the only
        # variable is the size fo each channel. here, they would be 7x7
        self.convout_size = 64 * 7 * 7
        self.classifier = nn.Sequential(
            nn.Linear(self.convout_size, 1024),
            nn.ReLU(inplace=True),
            nn.Dropout(p=0.25), # apply drop out to combat overfitting
            nn.Linear(1024, self.num_classes)
        )
        
    def forward(self, x):
        x = self.features_detector(x)
        x = x.view(x.size(0), self.convout_size)
        x = self.classifier(x)
        return x
    

# define the network
num_classes = 10
net = ConvNet(num_classes)

if is_cuda:
    net = net.cuda()

# define the loss function and the optimizer
# here we use common classification loss function: cross-entropy
criterion = nn.CrossEntropyLoss()
# we use stochastic gradient descent optimization
optimizer = torch.optim.SGD(net.parameters(), lr=learning_rate, momentum=momentum)
```

```python
# start training the network
net.train() # toggle the model into training mode
for epoch in range(training_epochs):
    training_loss = 0.0
    starttime = time.time()
    
    for i, data in enumerate(trainloader, 0):
        # unpack the features and the labels
        features, labels = data
        
        # wrap them in Variable
        if is_cuda:
            features, labels = Variable(features.cuda()), Variable(labels.cuda())
        else:
            features, labels = Variable(features), Variable(labels)
        
        # zero the parameter gradients
        optimizer.zero_grad()
        
        # feed forward the batch to the network and compute the loss
        output = net(features)
        loss = criterion(output, labels)
        
        # compute the gradients and update the weights
        loss.backward()
        optimizer.step()
        
        training_loss += loss.data[0]
        
    # report progress of loss of each epoch
    print ('[epoch: %d] loss: %.3f, elapsed time: %.2f' % 
        (epoch+1, training_loss / (i+1), time.time() - starttime))
```

```python
# measure the performance on test data
correct = 0
total = 0

mispredicted = []

net.eval() # toggle the model to testing mode
for data in testloader:
    # unpack the features and the labels
    features, labels = data
    
    # wrap them in Variable
    if is_cuda:
        features, labels = Variable(features.cuda()), labels.cuda()
    else:
        features = Variable(features)
        
    # feed forward the batch to the network and take the max
    output = net(features)
    _, predicted = torch.max(output.data, 1)
    
    total += labels.size(0)
    correct += (predicted == labels).sum()
    
    # record misclassified data
    for i in range(len(predicted)):
        if not predicted[i].cpu().numpy()[0] == labels[i]:
            mispredicted.append((features[i], labels[i], predicted[i]))
    
print ('Test accuracy: %.3f%%' % (100 * float(correct) / float(total)))
```

```python
# visualized the first 64 mistakes
# the format of the title [correct label] -> [predicted label]
plt.subplots_adjust(hspace=.8)
for i in range(64):
    sbplt = plt.subplot(8,8,1+i)
    sbplt.set_title('[' + str(mispredicted[i][1]) + '->' +
                    str(mispredicted[i][2].cpu().numpy()[0]) + ']') 
    sbplt.get_xaxis().set_visible(False)
    sbplt.get_yaxis().set_visible(False)
    sbplt.imshow(imgunloader(mispredicted[i][0].data.cpu()))
```
